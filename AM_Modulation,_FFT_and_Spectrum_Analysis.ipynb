```python
import numpy as np
import scipy.signal as sg
import matplotlib.pyplot as plt
import math
#from numpy.lib import twodim_base
from numpy.fft import rfft,fft,ifft

sample_rate = 48000
n_samples = 320
baseband_freq = 700
amplitude_signal = 0.5
amplitude_carrier = 1.0
carrier_freq = 8000
time = np.arange(n_samples) / sample_rate

baseband_signal = amplitude_signal * np.sin (2* math.pi * baseband_freq * time)
#plt.plot(baseband_signal)
#Baseband Signal ^^^^

modulated_signal = (amplitude_carrier + amplitude_signal*np.sin(2 * math.pi * baseband_freq * time)) * np.sin(2 * math.pi * carrier_freq * time)
#plt.plot(modulated_signal)
#Modulated Signal ^^^^

spectrum1 = abs(rfft(modulated_signal))
spectrum2 = np.abs(spectrum1)/1
spectrum2[1:-1] *= 2  # single-sided scaling
plt.plot(spectrum2)
plt.xlim(0, n_samples/2)
plt.locator_params(axis='x', nbins=20)
plt.axvline(x=53.033, ymax=1, color='red', linestyle='--')
#Spectrum ^^^^
```
