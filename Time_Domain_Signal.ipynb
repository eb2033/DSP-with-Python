```python
import numpy as np
import matplotlib.pyplot as plt

fs = 2000 # samples per second, increased for higher frequencies
T = 1.0 # time duration
N = int(fs*T) # number of samples
t = np.arange(N)/fs # array of samples

# Generate a signal with two distinct sine waves
x = 1.2*np.sin(2*np.pi*50*t) + 0.7*np.sin(2*np.pi*120*t)

X = np.fft.fft(x)
f = np.fft.fftfreq(N, d=1/fs)
A = np.abs(X)/N
A[1:-1] *= 2  # single-sided scaling

plt.figure()
plt.plot(t*1000, x)
plt.xlim(0, 50)  # ms, adjusted x-axis limit to show a few cycles of the combined wave
plt.xlabel('Time [ms]'); plt.ylabel('Amplitude'); plt.title('Combined Sine Waves (Time Domain)');

plt.figure()
plt.stem(f, A)
plt.xlim(0, 150) # Hz, adjusted x-axis limit to focus on the two peaks
plt.xlabel('Frequency [Hz]'); plt.ylabel('Amplitude [Vpeak]'); plt.title('Single-sided Amplitude Spectrum');

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

fs = 2000
T = 10
N = int(fs*T)
t = np.arange(N)/fs
#x = 1.2*np.sin(2*np.pi*50*t) + 0.7*np.sin(2*np.pi*120*t)
x = 1.2*signal.square(2*np.pi*5*t)

X = np.fft.fft(x)
f = np.fft.fftfreq(N, d=1/fs)
A = np.abs(X)/N
A[1:-1] *= 2  # single-sided scaling

plt.figure()
plt.plot(t*1000, x)
plt.xlim(0, 10)  # ms
plt.xlabel('Time [ms]'); plt.ylabel('Amplitude [V]'); plt.title('Time-domain');

plt.figure()
plt.stem(f, A)
plt.xlim(-1000, 1000)
plt.xlabel('Frequency [Hz]'); plt.ylabel('Amplitude [Vpeak]'); plt.title('Single-sided Amplitude Spectrum');

plt.show()
```
